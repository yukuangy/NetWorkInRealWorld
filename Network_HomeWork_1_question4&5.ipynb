```python

degrees = [degree for node, degree in G.degree()]


degree_freq = {}
for degree in degrees:
    if degree not in degree_freq:
        degree_freq[degree] = 0
    degree_freq[degree] += 1


num_nodes = len(G.nodes())
for degree in degree_freq:
    degree_freq[degree] /= num_nodes


plt.figure()
plt.loglog(list(degree_freq.keys()), list(degree_freq.values()), 'o')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution')
plt.show()
```

```python
import networkx as nx

G = nx.read_edgelist("facebook_combined.txt")
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
```

```python
import matplotlib.pyplot as plt

degrees = [degree for node, degree in G.degree()]
avg_degree = sum(degrees) / num_nodes


plt.hist(degrees, bins=50, edgecolor="black")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree Distribution")
plt.show()

print(f"Minimum degree: {min(degrees)}")
print(f"Maximum degree: {max(degrees)}")
print(f"Average degree: {avg_degree}")
```

```python
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
# Find the largest connected component (also called the giant component)
giant_component = max(nx.connected_components(G), key=len)
subgraph = G.subgraph(giant_component)

print(f"Size of the giant component: {len(giant_component)}")

# Compute diameter
diameter = nx.diameter(subgraph)
print(f"Diameter: {diameter}")
```

```python
vertex_connectivity = nx.node_connectivity(G)
edge_connectivity = nx.edge_connectivity(G)

print(f"Vertex connectivity: {vertex_connectivity}")
print(f"Edge connectivity: {edge_connectivity}")
```

```python
max_k = max(dict(G.degree()).values())
max_edge_conn = 0
k_for_max_edge_conn = 0

for k in range(1, max_k+1):
    k_core_subgraph = nx.k_core(G, k=k)
    if k_core_subgraph.number_of_nodes() == 0:  # Check if the k-core subgraph is empty
        continue
    edge_conn = nx.edge_connectivity(k_core_subgraph)
    if edge_conn > max_edge_conn:
        max_edge_conn = edge_conn
        k_for_max_edge_conn = k

print(f"Largest edge-connectivity using k-core: {max_edge_conn} for k={k_for_max_edge_conn}")
```

```python
largest_kcore = nx.k_core(G, k=k_for_max_edge_conn)
print(f"Number of nodes in largest k-core: {largest_kcore.number_of_nodes()}")
```

```python
vertex_conn_kcore = nx.node_connectivity(largest_kcore)
print(f"Vertex connectivity of the k-core: {vertex_conn_kcore}")
```

Describe in a few sentences a different approach that you could try to increase
the vertex connectivity of the network.

Ans: To increase the vertex connectivity of the network, one potential approach is to iteratively identify and remove the most vulnerable nodes or edges—those whose removal would lead to the disconnection of the graph. After each removal, we can create additional edges to connect the immediate neighbors of the removed node, ensuring that the graph remains connected. By continually strengthening the network in this manner, we make it more robust to node removal, thereby increasing its vertex connectivity. This approach seeks to minimize the graph's reliance on individual nodes or edges for maintaining connectivity, distributing the network's connectivity more uniformly across its nodes and edges.
